```python
import json
def OpenJsonFile(file) :
    with open(file, 'r') as file :
        data = json.loads(file.read())

    return data
    
def MergeData() :
    JsonList = ['eula.json', 'qcard.json']
    basedata = dict()
    
    for json in JsonList :
        data = OpenJsonFile(json)
        for cntrycode, value in data.items() :
            if cntrycode in basedata.keys() :
                for key in data[cntrycode].keys() :
                    if key.lower() not in basedata[cntrycode].keys() :
                        basedata[cntrycode][key] = data[cntrycode][key]
            else:
                basedata[cntrycode] = value

    return basedata
```

```python
d =  MergeData()
jd = json.dumps(d, indent=4)
print(jd)
```

```python
import pandas as pd
import json

def ParsingData(df) :
    return df['App'] + ':' + df['On/Off']

hotkey = pd.read_excel('Hot key.xlsx', sheet_name='RemoteControlMgmt_S23Y')
pubHotkey = hotkey[hotkey['Status'] == 'Publish'][['Country-Code', 'Country', 'App','On/Off']].reset_index(drop=True)
pubHotkey['app_id:isActive'] = pubHotkey.apply(ParsingData, axis=1)

pubHk = pubHotkey[['Country-Code', 'Country', 'app_id:isActive']]
pubDB = pubHk.groupby(['Country-Code','Country'], as_index=False).agg(lambda x : ', '.join(set(x))).set_index('Country-Code')

pubDB['app_id:isActive'] = pubDB['app_id:isActive'].map(lambda x: x.split(', '))
pubDB.to_json('hotkey.json', orient='index', indent=4)
```

```python
from collections import defaultdict 

hotkey = pd.read_excel('Hot key.xlsx',sheet_name='RemoteControlMgmt_S23Y')
pubHotkey = hotkey[hotkey['Status'] == 'Publish'][['Country-Code','Key', 'App', 'On/Off']].reset_index(drop=True)
pubHotkey = pubHotkey.rename({'App':'app_id', 'On/Off':'isActive'}, axis=1)
grouped = pubHotkey.groupby(['Country-Code'])

result = {}

for name, group in grouped :
    result[name[0]] = defaultdict(list)
    for i in range(len(group)) :
        result[name[0]][group.iloc[i]['Key']].append(
            {
                'app_id' : group.iloc[i]['app_id'],
                'isActive' : group.iloc[i]['isActive'],
            }
        )
        
with open('hotkey.json', 'w') as file :
    json.dump(result, file, indent=4)
    
```
